## Ejercicio 2:

1. Escriba una función que recibe una palabra en ingles y la traduce a "ubi-inglés" en tanto a cada vocal se le antepone "ub". Ejemplos:

    * octopus --> ubotubopuubs
    * soap -->suboubap
  

  
### Solución


```python
def ingles_ubi(palabra):
    salida=[]
    for letra in palabra:
        if letra in 'aeiou':
            salida.append(f'ub{letra}')
        else:
            salida.append(letra)
    return ''.join(salida)

print(ingles_ubi('sound'))
```

## Ejercicio 5:

Escribir una función 'suma_todo()' que reciba una cantidad arbitraria de elementos y los suma. Los elementos a sumar tienen que ser del mismo tipo y responder a '+' (por ejemplo enteros, flotantes, strings, o listas).

Por ejemplo: suma_todo('yo', 'soy', 'ufro') -> 'yosoyufro'; suma_todo([1,2,3], [4,5,6])->[1,2,3,4,5,6];
suma_todo(1,2,3)->6

```python
# Solución (similar a mysum!)

def suma_todo(*items):
    if not items:
        return items
    salida=items[0]
    for item in items[1:]:
        salida += item
    return salida

print(suma_todo())
print(suma_todo(10,20,30,40,50))
print(suma_todo('a', 'e', 'i', 'o', 'u'))
print(suma_todo([10,20,30], [40,50]))
```

## Ejercicio 7:

Crear una función que reciba una lista de strings y entregue el string de la lista que tiene el mayor número de letras repetidas.
Por ejemplo:  max_letras_repetidas(['tengo', 'acceso', 'a', 'mis', 'cuentas']) = 'acceso'

```python
## Clase Counter: Subclase de dict definida en la libreria collections que permite contar facilmente
from collections import Counter

print(Counter('absabsabbbc'))

# Puede entregar información sobre el orden de ocurrencia

print(Counter('absabsabbbc').most_common())

print(Counter('absabsabbbc').most_common(2))

def contador_de_letras_mas_repetidas(palabra):
    return Counter(palabra).most_common(1)[0][1]

contador_de_letras_mas_repetidas('absabsabbbc')
```

```python
# SOLUCIÓN (utilizando max con key, similar al caso sorted)
def contador_de_letras_mas_repetidas(palabra):
    return Counter(palabra).most_common(1)[0][1]

def max_extras_repetidas(lista_de_strings):
    return max(lista_de_strings, key=contador_de_letras_mas_repetidas)

LISTA_DE_PALABRAS=['tengo', 'acceso', 'a', 'mis', 'cuentas']

print(max_extras_repetidas(LISTA_DE_PALABRAS))
```

Tenemos la lista de estudiantes del curso con su número de matricula. Escriba una función que reciba la lista y la imprima ordenada por el número de matrícula con el siguiente formato:

Gomez     Walter    19283984

Villegas  Leonardo  83798039

O sea: nombre y apellidos con 10 caracteres, el número de matricula con 4 reducido a 4 caracteres

```python
NUMEROS_MATRICULA=[
 ('MONSERRATH', 'ALVAREZ', '20354298418'),
 ('CAMILA', 'CARTES', '20985253520'),
 ('MARCELO', 'CASTRO', '21073339621'),
 ('ARELY', 'CONCHA', '21021862920'),
 ('LUIS', 'GUERRERO', '20357299919'),
 ('DANIEL', 'KOH', '20638424719'),
 ('GERMÁN', 'LAGOS', '20919321320'),
 ('ALAN', 'LÓPEZ', '20532934K20'),
 ('JAVIER', 'LUENGO', '21089099821'),
 ('JOSÉ', 'MATUS', '21266849421'),
 ('NICOLÁS', 'MONTECINOS', '19957230k17'),
 ('IGNACIO', 'MORALES', '20081365019'),
 ('STEVEN', 'NAHUELPÁN', '20492813419'),
 ('JAVIERA', 'OLAVE', '19320209814'),
 ('SEBASTIÁN', 'OLIVA', '20147867718'),
 ('MANLET', 'SÁNCHEZ', '20367020619'),
 ('MILLARAY', 'TRANGOL', '20872536K20'),
 ('CAROLINA', 'VALLEJOS', '18720137319')
]
```

```python
#Solución

import operator

def lista_ordenada_y_formateada(lista_de_tuplas):
    salida=[]
    formato='{1:20} {0:20} {2:4}'
    for persona in sorted(lista_de_tuplas, key=operator.itemgetter(2)):
        salida.append(formato.format(*persona))
    return salida

print('\n'.join(lista_ordenada_y_formateada(NUMEROS_MATRICULA)))
```
