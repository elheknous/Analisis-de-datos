```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetro de la distribución de Poisson (lambda)
mu = 5

# Valores posibles (discretos)
x = np.arange(0, 20)

# Probabilidades de la distribución de Poisson
pmf = poisson.pmf(x, mu)

# Crear el gráfico
plt.bar(x, pmf, color='skyblue')
plt.xlabel('Número de eventos')
plt.ylabel('Probabilidad')
plt.title('Distribución de Poisson (mu = {})'.format(mu))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, gamma, expon

# Configuración del gráfico
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

# Distribución Beta
a, b = 2, 5  # Parámetros de forma
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)
axs[0].plot(x, y, 'r-', lw=2, label='beta pdf')
axs[0].set_title('Distribución Beta')
axs[0].set_xlabel('x')
axs[0].set_ylabel('Densidad de probabilidad')
axs[0].legend()

# Distribución Gamma
shape, scale = 2, 2  # Parámetros de forma y escala
x = np.linspace(0, 20, 1000)
y = gamma.pdf(x, shape, scale=scale)
axs[1].plot(x, y, 'b-', lw=2, label='gamma pdf')
axs[1].set_title('Distribución Gamma')
axs[1].set_xlabel('x')
axs[1].set_ylabel('Densidad de probabilidad')
axs[1].legend()

# Distribución Exponencial
scale = 1  # Parámetro de escala
x = np.linspace(0, 10, 1000)
y = expon.pdf(x, scale=scale)
axs[2].plot(x, y, 'g-', lw=2, label='exponencial pdf')
axs[2].set_title('Distribución Exponencial')
axs[2].set_xlabel('x')
axs[2].set_ylabel('Densidad de probabilidad')
axs[2].legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
