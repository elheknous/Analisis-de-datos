```python
df = pd.read_csv(r"comunas.csv", encoding='latin1')
df
```

```python
data = df['IDC']
data
```

```python
#BETA

from scipy.stats import beta, kstest

a, b = 2, 5

d, p_value = kstest(data, 'beta', args=(a, b))

print(f'Estadístico D: {d}')
print(f'Valor p: {p_value}')
```

```python
import numpy as np
from scipy.stats import beta, kstest

# Supongamos que 'sample' es tu conjunto de datos
# Por ejemplo, podrías cargar tus datos desde un archivo o definirlos directamente
sample = data  # Esta línea es solo un ejemplo. Reemplázala con tus datos.

# Estimar los parámetros de la distribución beta que mejor se ajustan a tus datos
a, b, loc, scale = beta.fit(sample)

# Realizar el test de Kolmogorov-Smirnov para la distribución beta
d, p_value = kstest(sample, 'beta', args=(a, loc, scale))

print(f'Estadístico D: {d}')
print(f'Valor p: {p_value}')

# Interpretar el resultado
if p_value > 0.05:
    print("No se puede rechazar la hipótesis nula. La muestra podría seguir una distribución beta.")
else:
    print("Se rechaza la hipótesis nula. La muestra no sigue una distribución beta.")

```
