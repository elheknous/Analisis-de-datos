La creación de vectores se realiza bajo la funcción linspace(), es muy útil para representar intervalos $[a,b]$, los argumentos básicos son de tipo numérico (float o int):
-   start: inicio del intervalo cerrado (obligatorio)
-   stop: fin del intervalo cerrado (obligatorio)
-   num: número de elementos que se registrarán, el valor por defecto es 50 y si se omite el argumento se reemplazará por este valor.

Existen más argumentos como dtype(por defecto numpy.float) que especifica el tipo de datos resultante en el vector.

Otra función importante es meshgrid(), necesaria para describir una región en dos dimensiones. Sus argumentos elementales son:
-   x: Un vector que representa al eje x en la región de interés.
-   y: Un vector que representa al eje y en a misma región. 


```python
import numpy as np

# Ejemplos de linspace()

X = np.linspace(start=0,stop=1,num=11)  # Vector de numpy, inicia en 0, termina en 1 y tiene 11 elementos.
print('np.linspace(0,1,11)= ',X)  # Es equivalente escribir omitiendo el nombre de los argumentos.

X = np.linspace(-10,10,21) # inicio en -10, fin en 10, 21 elementos.
print('\nnp.linspace(-10,10,20)= ',X)

X = np.linspace(0,49) # Aquí el arugmento num se omite y toma su valor por defecto(50)
print('\nnp.linspace(0,49)= ',X)

X = np.linspace(0,49, dtype=int) # vector de enteros.
print('\nnp.linspace(0,49)= ',X)
```

```python
#  Ejemplo de meshgrid()

x = np.linspace(0,9,10,dtype=int)  # Un vector para el eje x en [0,9]
y = np.linspace(10,19,10,dtype=int) # Un vector para el eje y en [10,19]

X,Y = np.meshgrid(x,y)  # Obtenemos dos vectores compuestos por vectores
                        # Estos describen un producto cartesiano x X y

print('Eje x: ',x)
print('Eje y: ',y)

print(X,Y)
a = 6  # <- control del eje y
b = 9  # <- control del eje x
X[a][b], Y[a][b]  # Puede describir las coordenadas (x,y)
```

# Ejemplo

Veamos una aplicación. La función de densidad conjunta de X,Y es dada por:

$$
f(x, y) = \begin{cases} 
\frac{2}{7} (3x + 4y) & \text{si } 0 < x < 1 , 0 < y < 1 \\ 
0 & \text{en otro caso} 
\end{cases}
$$


```python
# Definiendo un dominio y un recorrido para graficar. 

X = np.linspace(start=0,stop=1,num=101)  # Define un intervalo en [0,1] para representar a x.
print('Vector definido para representar a X: \n',X)

Y = np.linspace(0,1,101)  # Si bien en este caso podría utilizar el mismo vector anterior, agreguemos otro intervalo [0,1] para representar a y.
print('\nVector definido para representar a Y: \n',Y)

X,Y = np.meshgrid(X,Y)  # Espacio bidimensional generado por X e Y

Z = 2/7 * (3*X + 4*Y) # Este resultado será una matriz, con todas las imágenes de f(x,y)
print('\nImágenes Z = f(x,y) para todo x en X e y en Y: \n',Z)

print(f'\nf({X[0][2]} , {Y[0][2]}) = ',Z[0][2])
 
```

3- Graficar con matplotlib es muy sencillo cuando ya tiene definido el dominio y el recorrido de su función.

```python
import matplotlib.pyplot as plt

X = np.linspace(start=0,stop=1,num=101)
Y = np.linspace(0,1,101)
X,Y = np.meshgrid(X,Y) 
Z = 2/7 * (3*X + 4*Y)


fig = plt.figure()  # Primero se debe crear una figura (es como la planilla en blanco)
ax = fig.add_subplot(projection='3d')  # luego se debe agregar un gráfico, el argumento "projection" indica que será en 3 dimensiones

ax.plot_surface(X,Y,Z,cmap='cividis')  # Agregar los datos X,Y que forman el meshgrid, y Z las imágenes.
ax.set_aspect('equal')  # Relación de aspecto de los ejes.
ax.set_xlabel('X')  # Nombre para el eje x
ax.set_ylabel('Y')   # Nombre para el eje y
ax.set_zlabel('f(x,y)') # Nombre para el eje z
plt.title('Gráfico 1') # Título de la gráfica

plt.show()  # Mostrar gráfico.

 # Para crear un gráfico interactivo es mejor ejecutar un script .py en lugar de jupyter notebook.
```

Ejemplo 2:
A veces las funciones tienen condiciones más complicadas y las imágenes deben calcularse de otra forma.
Sea la función de densidad conjunta de (X,Y) dada por:

$$
f(x, y) = \begin{cases} 
8xy & \text{si } 0 < x < y < 1  \\ 
0 & \text{en otro caso} 
\end{cases}
$$

Para ello es necesario utilizar la función where() de numpy, su objetivo es filtrar matrices bajo sentencias lógicas. Los argumentos van en la siguiente forma:
-   Salida = np.where(P,R, nR)

-   P = condición lógica
-   R = resultado si la condición es verdadera
-   nR= resultado si la condición es falsa


En numpy el operador lógico "and" se representa con un "&", mientras que "or" utiliza "|" y la negación "~" .


```python
# Los pasos son casi los mismos.

X = np.linspace(0,1,100)
Y = np.linspace(0,1,100)

X,Y = np.meshgrid(X,Y)

Z = np.where((0<X) & (X<Y) & (Y<1), 8*X*Y,0)  # Donde la condición (0<X) & (X<Y) & (Y<1) sea cierta, aplica 8*x*Y, de lo contrario aplica el valor 0

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap='cividis')
ax.set_aspect('auto')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
plt.title('Gráfico 2')

plt.show()
```

```python
X = np.linspace(0,1,100)
Y = np.linspace(0,1,100)
X,Y = np.meshgrid(X,Y)
Z = np.where((0<X) & (X<Y) & (Y<1), 8*X*Y,0) 
# Máscara para marcar superficies específicas:


fig = plt.figure()
ax = fig.add_subplot(projection='3d')
surperficie= ax.plot_surface(X,Y,Z,cmap='cividis',edgecolor='none',alpha=0.7)

mask = (X > 0.4) & (X < 0.6) & (X < Y)  # por ejemplo marcar donde se cumpla la condición
z_mask = np.where(mask, Z, np.nan) 

ax.plot_surface(X,Y,z_mask, cmap='autumn',edgecolor='red')
ax.plot_wireframe(X,Y,z_mask, color='red')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x,y)')
plt.title('Gráfico 3')

plt.show()
```
