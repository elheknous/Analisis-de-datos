```python
x = np.linspace(-2*np.pi,2*np.pi,200)

plt.plot(x, np.sin(x), color='red')
plt.title('Función Seno de x')
plt.grid()
```

```python
x = np.linspace(-2*np.pi,2*np.pi,200)

plt.plot(x,np.sin(x), label='curva continua')
samples = x[::4]
plt.plot(samples,np.sin(samples),'r*', label='valores discretos')
plt.title('Función Seno de x')
plt.grid()
plt.legend(loc='upper right', fontsize='small')
```

**Gráfico de Dispersión**

```python
x1 = 2*np.random.standard_normal((2,100))
x2 = 0.8*np.random.standard_normal((2,100)) + np.array([[6],[2]])

plt.plot(x1[0],x1[1],'s')
plt.plot(x2[0],x2[1],'r*')
plt.title('2D scatter plot')
```

**Histograma**

```python
sigma, mu = 2, 10
x = sigma*np.random.standard_normal(10000)+mu

plt.hist(x, 50, density=True, facecolor='b', alpha=0.75, histtype='step')
z = np.linspace(0,20,200)

plt.plot(z, (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-(z-mu)**2/(2*sigma**2)),'r')
plt.title('Histograma')
```

**Gráfica de Tortas**

```python
chile = [1024, 2334]
arg = [1331, 2760]

peru = [791, 2567]
brasil = [1009, 3082]

explode = (0, 0.1) 

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0,0].pie(chile, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
axs[0,0].axis('equal')
axs[0,0].set_title("Chile")

axs[0,1].pie(arg, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
axs[0,1].axis('equal')
axs[0,1].set_title("Argentina")

axs[1,0].pie(peru, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
axs[1,0].axis('equal')
axs[1,0].set_title("Perú")

axs[1,1].pie(brasil, explode=explode, autopct='%1.1f%%',
        shadow=True, startangle=90)
axs[1,1].axis('equal')
axs[1,1].set_title("Brasil")

fig.legend(['Negative', 'Positive'], loc='lower center')
plt.show()
```
