1. Cree un diccionario en el cual se tengan las claves como nombre de usuarios y los valores como passwords, ambos representados como string. Escriba una función login. El usuario debe ingresar su nombre de usuario y password, y el sistema le indica si ha logrado acceder, o sino le rechaza el acceso.

```python
def login():
  mi_diccionario = {'jose': '123', 'juliana': 'lacra123', 'espe': 'bussyban'}
  usuario = input("Ingrese su usuario: ")
  if usuario in mi_diccionario:
    while True:
      clave = input("Ingrese su clave: ")
      if clave == mi_diccionario[usuario]:
        print("Ha accedido con exito")
        break
      else:
        print("Clave incorrecta, intente de nuevo")
  else:
    print("Usuario no encontrado")
login()
```

3. Cree un diccionario cuyas claves sean los nombres de las personas de su familia, y cuyos valores sean las fechas de nacimiento. Escriba ahora una función que reciba el nombre de alguien de su familia e imprima la edad actual de la persona en días.

```python
from datetime import datetime

def cumple(nombre):
  fam = {"jose": "26/03/2003", "mama": "03/10/1970", "ale": "02/04/2002"}
  if nombre in fam:
    dias = calcular_dias_de_vida(fam[nombre])
    return dias
  else:
    return "No esta es nombre"



def calcular_dias_de_vida(fecha_nacimiento_str):
    # Convertir la cadena de texto a un objeto de fecha
    fecha_nacimiento = datetime.strptime(fecha_nacimiento_str, "%d/%m/%Y")

    # Obtener la fecha actual
    fecha_actual = datetime.now()

    # Calcular la diferencia entre la fecha actual y la fecha de nacimiento
    diferencia = fecha_actual - fecha_nacimiento

    # Extraer el número de días de la diferencia
    dias_de_vida = diferencia.days

    return dias_de_vida


print(cumple("jose"))
```
