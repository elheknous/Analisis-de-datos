Escriba una versión alternativa de la función ingles_secreto() que considere las siguientes reglas:

```python
def ingles_secreto(palabra):
  contador = 0
  cos = "aeiou"
  for i in palabra:
    if i in cos:
      contador = contador + 1

  if contador > 1:
      return f'{palabra}way'
  return f'{palabra[1:]}{palabra[0]}ay'

print(ingles_secreto("wine"))
```

Modifique el traductor ubi_ingles() (clase Ejercicios 02 - Ejercicio 2) para que trabaje correctamente las palabras capitalizadas, o sea si recibe una palabra capitalizada, debe entregar la traducción a ubi-ingles también capitalizada.

```python
def ingles_ubi(palabraaa):
  if palabraaa.istitle():
    palabra = palabraaa.lower()
    salida=[]
    for letra in palabra:
        if letra in 'aeiou':
            salida.append(f'ub{letra}')
        else:
            salida.append(letra)
    return ''.join(salida).capitalize()
  else:
    salida=[]
    for letra in palabraaa:
        if letra in 'aeiou':
            salida.append(f'ub{letra}')
        else:
            salida.append(letra)
    return ''.join(salida)


print(ingles_ubi('Sound'))
```

Crear una función mas_menos() que reciba una lista o tupla de números y devuelva el resultado de sumar/restar alternativamente los números. Por ejemplo: mas_menos([[10, 30, 50, 60, 70]])-> 40 (=10-30+50-60+70)



```python
lista = [10, 30, 50, 60, 70]
def mas_menos(lista):
  suma = 0
  for i in range(len(lista)):
    if i % 2 == 0:
      suma = suma + lista[i]
    else:
      suma = suma - lista[i]
  return suma
print(mas_menos(lista))
```

Implemente una propia función mi_zip() que recibe una cantidad indeterminada de iterables y entrega una lista de tuplas, donde cada tupla contiene un elemento de los iterables ingresados. (Hint: Puede entregar una lista (no un iterable) de vuelta y asumir que todos los iterables ingresados tienen el mismo largo):

Ejemplo: mi_zip('abc',[1,2,3], ['A', 'B', 'C']) --> [('a',1,'A'),('b',1,'B'),('c',1,'C')]

```python
def mi_zip(*iterables):
    longitud_minima = min(len(iterable) for iterable in iterables)
    resultado = []
    for i in range(longitud_minima):
        tupla = tuple(iterable[i] for iterable in iterables)
        resultado.append(tupla)
    return resultado

# Ejemplo de uso:
resultado = mi_zip('abc', [1, 2, 3], ['A', 'B', 'C'])
print(resultado)  # Output: [('a', 1, 'A'), ('b', 2, 'B'), ('c', 3, 'C')]
```

Implementar una función suma_numerica() que recibe una cantidad arbitraria de argumentos. Si un argumento es entero, o puede ser interpretado como tal, entonces se suma al valor total. En caso contratrio se ignora el argumento correspondiente.

Ejemplo: suma_numerica(10,20,'casa','30') --> 60=10+20+30 (notar que '30' se interpreta como entero aunque es un string)

```python
def suma(*numeros):
  suma = 0
  for i in numeros:
    if type(i) == int or type(i) == float:
      suma += i
    else:
      try:
        suma += int(i)
      except ValueError:
        pass
  return suma

print(suma(10,20,'casa','30'))
```

Cree una función que reciba una lista de diccionarios y devuelva un único diccionario que combina todos las claves y valores. Si una clave aparece en mas de un diccionario, el correspondiente valor debe ser una lista de todos los valores asociados.

Ejemplo: combina_diccionarios([{'clave1':10, 'clave2':20}, {'clave2':30, 'clave3':40}]) --> {'clave1':10, 'clave2':[20,30], 'clave3':40}

```python
def combina_diccionarios(lista_diccionarios):
    resultado = {}

    for diccionario in lista_diccionarios:
        for clave, valor in diccionario.items():
            if clave in resultado:
                if isinstance(resultado[clave], list):
                    resultado[clave].append(valor)
                else:
                    resultado[clave] = [resultado[clave], valor]
            else:
                resultado[clave] = valor

    return resultado

# Ejemplo de uso
lista = [{'clave1':10, 'clave2':20}, {'clave2':30, 'clave3':40}]
print(combina_diccionarios(lista))  # Salida: {'clave1': 10, 'clave2': [20, 30], 'clave3': 40}
```
